np.meshgrid(x)[0][0] => 1



Make N-D coordinate arrays for vectorized evaluations of N-D scalar/vector fields over N-D grids

### contour
contour and contourf draw contour lines and filled contours, respectively

contour([X, Y,] Z, [levels], **kwargs)

X, Y : array-like

X and Y must both be 2-D with the same shape as Z (e.g. created via numpy.meshgrid), 

or 

they must both be 1-D such that len(X) == M is the number of columns in Z and len(Y) == N is the number of rows in Z.

像坐標軸

Z: The height values over which the contour is drawn.

### plt.axes().set_aspect('equal')
https://jdhao.github.io/2017/06/03/change-aspect-ratio-in-mpl/

    axes 對稱軸
    set_aspect: set the aspect ratio(長寬比)
    0.5 ==> x 軸上一單位的長度，僅為 y 0.5 單位的長度
        也就是，x 軸的單位長度較長，若一格 5 ，y 同樣的一格則代表 10，但長度一樣
        所以想要讓 x 軸的格子分的細一點，可以加劇 x 的變化，就把比例條小

```python
import math

plt.figure()
xlist = np.linspace(0, 1, 100) # only in quadrant I
ylist = np.linspace(0, 1, 100)
X,Y = np.meshgrid(xlist, ylist)
xlist1 = np.linspace(1, 2, 100) # only in quadrant I
ylist1 = np.linspace(1, 2, 100)
X1,Y1 = np.meshgrid(xlist1, ylist1)
F = X**2 + Y**2 - 1  #  'Circle Equation
plt.contour(X, Y, F, [0], colors = 'k', linestyles = 'solid')

plt.contour(X1, Y1, F, [0], colors = 'k', linestyles = 'solid')
plt.axes().set_aspect('equal')

plt.show()
# https://stackoverflow.com/questions/58063554/how-to-draw-quarter-circle-curves-in-python
```

```python
import math

def draw_quarters1(x_start, x_end, y_start, y_end, lower_right_quarter, upper_left_quarter):
    

    x_axis = np.linspace(x_start, x_end, 100) # only in quadrant I
    y_axis = np.linspace(y_start, y_end, 100)
    axis_x,axis_y = np.meshgrid(x_axis, y_axis)
    plt.contour(axis_x, axis_y, upper_left_quarter, [0], colors = 'k', linestyles = 'solid')
    plt.contour(axis_x, axis_y, lower_right_quarter, [0], colors = 'k', linestyles = 'solid')
    
def draw_quarters2(x_start, x_end, y_start, y_end, lower_right_quarter, upper_left_quarter):
    

    x_axis = np.linspace(x_start, x_end, 100) # only in quadrant I
    y_axis = np.linspace(y_start, y_end, 100)
    axis_x,axis_y = np.meshgrid(x_axis, y_axis)
    plt.contour(axis_x, axis_y, lower_right_quarter, [0], colors = 'k', linestyles = 'solid')
    plt.contour(axis_x, axis_y, upper_left_quarter, [0], colors = 'k', linestyles = 'solid')

plt.figure() #開始畫圖

xlist1 = np.linspace(0, 1, 100) # 畫右上角這個 1/4 圓
ylist1 = np.linspace(0, 1, 100)
X1,Y1 = np.meshgrid(xlist1, ylist1)

upper_right_quarter = X1**2 + Y1**2 - 1  #  'Circle Equation

xlist2 = np.linspace(-1, 0, 100) # 畫左下角這個 1/4 圓
ylist2 = np.linspace(-1, 0, 100)
X2,Y2 = np.meshgrid(xlist2, ylist2)
lower_left_quarter = X2**2 + Y2**2 - 1   

# xlist3 = np.linspace(1, 0, 100) # 畫右下角這個 1/4 圓
# ylist3 = np.linspace(0, -1, 100)
xlist3 = np.linspace(0, 1, 100) # 畫右下角這個 1/4 圓
ylist3 = np.linspace(-1, 0, 100)
X3, Y3 = np.meshgrid(xlist3, ylist3)
lower_right_quarter = X3**2 + Y3**2 - 1  

xlist4 = np.linspace(-1, 0, 100) # 畫左上角這個 1/4 圓
ylist4 = np.linspace(0, 1, 100)
X4, Y4 = np.meshgrid(xlist4, ylist4)
upper_left_quarter = X4**2 + Y4**2 - 1  



draw_quarters1(0, 1, 0, 1, upper_right_quarter, lower_left_quarter)
draw_quarters2(1, 2, 0, 1, lower_right_quarter, upper_left_quarter)

plt.axes().set_aspect('equal')

plt.show()
```

```python
for i in range(1, 6):
    for j in range(1, 6):
        # 偶數列先畫左上到右下的
        if i%2 == 0:
                
            draw_quarters1(2*j-1, 2*j, i, i+1, upper_right_quarter, lower_left_quarter)
            draw_quarters2(2*j, 2*j+1, i, i+1, lower_right_quarter, upper_left_quarter)
        
        # 奇數列先畫右上到左下的
        else:
            draw_quarters1(2*j, 2*j+1, i, i+1, upper_right_quarter, lower_left_quarter)
            draw_quarters2(2*j-1, 2*j, i, i+1, lower_right_quarter, upper_left_quarter)

            
plt.axes().set_aspect('equal')
```
