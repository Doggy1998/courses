```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
## 模擬 svm 資料
# 轉乘 np array
x = np.array([[6, 6], [10, 8], [3, 7], [8, 7]])
```

```python
x[:, 0]
```

```python
plt.scatter(x[:, 0], x[:, 1])
# 將 3,7 6,6 分一類
```

```python
y = np.array([1, 2, 1, 2])
```

```python
plt.scatter(x[:, 0], x[:, 1], c=y, cmap="Paired")
```

```python
plt.scatter(x[:, 0], x[:, 1], c=y, cmap="Set2", s=150)
```

```python
# 打開函數學習機
from sklearn.svm import SVC 
```

```python
clf = SVC(gamma='auto')
```

```python
clf.fit(x, y)
# gamma 控制資料解釋的精確度 若精確度過高 有可能 overfitted 預測新的資料能力可能較弱 
```

```python
y_predict = clf.predict(x)
```

```python
y_predict
```

```python
xr = np.random.rand(100)*7 + 3
yr = np.random.rand(100)*2 + 6
```

```python
plt.scatter(xr, yr)
plt.scatter(x[:, 0], x[:, 1], c=y, cmap="Paired", s=150)
```

```python
print('X', xr[:5])
print('Y', yr[:5])
```

```python
### 合併
X = np.c_[xr, yr]
```

```python
Y = clf.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap="Paired")
```

```python
### meshgrid
```

```python
x_list = np.linspace(3, 10, 50)
y_list = np.linspace(6, 8, 50)
X, Y = np.meshgrid(x_list, y_list)
P = np.c_[X.ravel(), Y.ravel()]
C = clf.predict(P)
plt.scatter(P[:, 0], P[:, 1], c=C)
```
