# #
#發現資料有 2013 2015 年的
#且全區資料不應該與其他國家一起分析
#各國家的也有 合計 或 小計 的資料，導致個國家總合大於 全區；應該刪除


```python
# 全區的資料 另外處理
countries = df[df.居住地 != '全區']
countries = countries[~df['細分'].str.contains("合計")]
countries = countries[~df['細分'].str.contains("小計")]
all2013 = df2013[df.居住地 == '全區']
all2015 = df2015[df.居住地 == '全區']

countries2013 = countries[(countries.年別==2013)]
countries2015 = countries[(countries.年別==2015)]
```
