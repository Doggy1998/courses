```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('https://quality.data.gov.tw/dq_download_csv.php?nid=9621&md5_url=8ae0214f8573ae27d20b890e984bfe55', encoding = 'utf-8')
```

```python
df.head()
```

```python
set(df.等級別)
```

```python
bachelor = df[(df.等級別=="B 學士")]
master = df[(df.等級別=="'M 碩士")]
doctor = df[(df.等級別=="D 博士")]
```

```python
bachelor.head()
```

```python
bCity = bachelor.groupby(['縣市名稱'])[["學生數", "教師數", '上學年度畢業生數']].sum()
```

```python
bCity
```

```python
bCity.sort_values(by=['學生數'], ascending=False).head()
```

```python
bCity.sort_values(by=['學生數'], ascending=False).tail()
```

```python
CitySchool = bachelor[["縣市名稱", "學校名稱"]]
```

```python
citySchool = CitySchool.drop_duplicates()
citySchool.head()
```

```python
citySchool.groupby(["縣市名稱"])[["學校名稱"]].count()
```

```python
citySchoolCount = citySchool.groupby(["縣市名稱"])[["學校名稱"]].count()
```

```python
mix = pd.merge(bCity, citySchoolCount, on=['縣市名稱'], )
```

```python
mix
```

```python
mix['師生比'] = mix['學生數']/mix['教師數']
```

```python
mix
```

```python
mix['學生數/學校數'] = mix['學生數']/mix['學校名稱']
```

```python
mix
```

```python
mix['上學年度畢業生數/學生數'] = mix['上學年度畢業生數']/mix['學生數']
```

```python
mix.sort_values(by=['學生數/學校數'], ascending=False)
```

## 觀察各縣市的 師生比 學生數/學校 上學年畢業生數/學生數 後
## 畫出 師生比 與 畢業率 的圖

```python
x = mix['師生比']
y = mix['上學年度畢業生數/學生數']

# Plot
plt.scatter(x, y)
plt.title('師生比 與 畢業率 的關係')
plt.xlabel('師生比')
plt.ylabel('上學年度畢業生數/學生數')
plt.show()
```
