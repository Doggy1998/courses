## 記帳
### 先輸入日期
### 可以設定當月預算
### 輸入支出項目及金額
### 可以輸出 txt 當日記帳結果

```python

budget = {}
expense = {}

# 依照每月每日 建立空白的 dictionary 
# 之後每一天的支出是以 {'item': "item1", "money": 100} 的形式儲存(append)在當日的 list 中
def initializeExpense(expense):
    for i in ['1', '3', '5', '7', '8', '10', '11', '12']:
        expense[i] = {}
        for j in range(1, 32):
            expense[i][str(j)] = []
    for i in ['4', '6', '9', '11']:
        expense[i] = {}
        for j in range(1, 31):
            expense[i][str(j)] = []
    expense['2'] = {}
    for j in range(1, 29):
        expense['2'][str(j)] = []
    return expense

# 獲得當月累積的支出
def getExpenseOfMonth(month):
    totalExpense = 0
    for day in expense[month]:
        totalExpense = totalExpense + sum([x['money'] for x in expense[month][day]])
    return totalExpense

# 將某一天的支出轉為字串輸出
def makeReport(month, day, expense):
    result = f'{month}/{day} 支出表\n'
    totalExpenseOfDay = 0
    for i in expense[month][day]:
        result += f"{i['item']}\t{i['money']}\n"
        totalExpenseOfDay += i['money']
    result += "="*50
    
    result += f"\nTotal: {totalExpenseOfDay}"
    return result


while True:
    # 空白每月每日 dictionary
    expense = initializeExpense(expense)
    
    month = input("請輸入月份 (若要結束輸入 0)")
    if month == "0":
        break
    day = input("請輸入日期")
    
    # 沒有當月預算可以設定
    if not budget.get(month):
        money = input(f"請輸入{month}月預算")
        budget[month] = int(money)
        
    # 已有當月預算也可以更改
    else:
        change = input(f"需要更改預算嗎?目前{month}月預算為{budget[month]}\nN: 維持即可 Y: 我要換")
        if change == "Y":
            money = input("請輸入新的{month}月預算")
            budget[month] = int(money)
    
    while True:
        function = input(f"A: 紀錄支出項目及金額\nB: 輸出歷史紀錄\nq: 結束{month}/{day}的支出紀錄")
        
        #紀錄支出
        if function == "A":
            item = input("請輸入支出項目: ")
            money = input("請輸入支出金額: ")
            expense[month][day].append({"item": item, "money": int(money)})
            monthExpense = getExpenseOfMonth(month)
            print(f"你{month}月還可以花{budget[month]-monthExpense}")
        
        # 輸出當日支出項目及金額
        elif function == "B":
            with open(month+day+".txt", "w+", encoding="utf-8") as file:
                report = makeReport(month, day, expense)
                file.write(report)
                print("輸出成功")
                
        elif function == "q":
            break
            
        else:
            print("請輸入正確的代號")
    
    
```
